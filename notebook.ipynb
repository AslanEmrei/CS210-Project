```python
spotify = AslanEmreIşık_SpotifyData.copy()
```

```python
# Convert 'endTime' to datetime if it's not already
spotify['endTime'] = pd.to_datetime(spotify['endTime'])

# Extract the date only (day-month-year)
spotify['date'] = spotify['endTime'].dt.date

# Group by the extracted date and sum up 'msPlayed'
spotify_day_time = spotify.groupby('date', as_index=False)['msPlayed'].sum()

spotify_day_time['date'] = pd.to_datetime(spotify_day_time['date'])
spotify_day_time.set_index('date', inplace=True)


# Convert 'msPlayed' from milliseconds to minutes
spotify_day_time['dailyListeningTime'] = spotify_day_time['msPlayed'] / 60000

# Drop the 'msPlayed' column if you only want the date and dailyListeningTime
spotify_day_time = spotify_day_time.drop(columns=['msPlayed'])


# Display the resulting DataFrame
print(spotify_day_time)
```

```python
max(spotify_day_time['dailyListeningTime'])
```

```python
import matplotlib.pyplot as plt

# Plot 1: Daily Listening Time Over Time
plt.figure(figsize=(12, 6))
plt.plot(spotify_day_time.index, spotify_day_time["dailyListeningTime"], marker='o')
plt.title("Daily Listening Time Over Time")
plt.xlabel("Date")
plt.ylabel("Daily Listening Time (minutes)")
plt.grid()
plt.show()

# Plot 2: Distribution of Daily Listening Time (Histogram)
plt.figure(figsize=(12, 6))
plt.hist(spotify_day_time["dailyListeningTime"], bins=20, edgecolor='black', color='skyblue')
plt.title("Distribution of Daily Listening Time")
plt.xlabel("Daily Listening Time (minutes)")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# Plot 3: Weekly Total Listening Time
weekly_total = spotify_day_time["dailyListeningTime"].resample("W").sum()
plt.figure(figsize=(12, 6))
plt.plot(weekly_total.index, weekly_total, marker='o', linestyle='-')
plt.title("Weekly Total Listening Time")
plt.xlabel("Week")
plt.ylabel("Total Listening Time (minutes)")
plt.grid()
plt.show()

# Plot 4: Monthly Total Listening Time
monthly_total = spotify_day_time["dailyListeningTime"].resample("M").sum()
plt.figure(figsize=(12, 6))
plt.bar(monthly_total.index.strftime('%Y-%m'), monthly_total, color='coral')
plt.title("Monthly Total Listening Time")
plt.xlabel("Month")
plt.ylabel("Total Listening Time (minutes)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot 5: Boxplot of Daily Listening Time
plt.figure(figsize=(8, 6))
plt.boxplot(spotify_day_time["dailyListeningTime"], vert=False, patch_artist=True)
plt.title("Boxplot of Daily Listening Time")
plt.xlabel("Daily Listening Time (minutes)")
plt.grid(axis='x')
plt.show()

# Plot 6: Rolling Average (7-day) of Daily Listening Time
rolling_avg = spotify_day_time["dailyListeningTime"].rolling(window=7).mean()
plt.figure(figsize=(12, 6))
plt.plot(spotify_day_time.index, rolling_avg, color='green')
plt.title("7-Day Rolling Average of Daily Listening Time")
plt.xlabel("Date")
plt.ylabel("Rolling Average (minutes)")
plt.grid()
plt.show()


# Plot 8: Listening Time by Day of the Week
spotify_day_time['dayOfWeek'] = spotify_day_time.index.day_name()
listening_by_day = spotify_day_time.groupby("dayOfWeek")["dailyListeningTime"].mean().reindex(
    ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
)
plt.figure(figsize=(12, 6))
plt.bar(listening_by_day.index, listening_by_day, color='teal')
plt.title("Average Listening Time by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Average Listening Time (minutes)")
plt.grid(axis='y')
plt.show()

# Plot 9: Heatmap of Daily Listening Time by Month and Day
import seaborn as sns
spotify_day_time['month'] = spotify_day_time.index.month
spotify_day_time['day'] = spotify_day_time.index.day
heatmap_data = spotify_day_time.pivot_table(
    index='day', columns='month', values='dailyListeningTime', aggfunc='sum'
)
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=False, fmt=".1f", cbar=True)
plt.title("Heatmap of Daily Listening Time by Month and Day")
plt.xlabel("Month")
plt.ylabel("Day")
plt.show()

# Plot 10: Listening Time Trends by Quarter
spotify_day_time["quarter"] = spotify_day_time.index.to_period("Q")
listening_by_quarter = spotify_day_time.groupby("quarter")["dailyListeningTime"].sum()
plt.figure(figsize=(12, 6))
plt.bar(listening_by_quarter.index.astype(str), listening_by_quarter, color='orange')
plt.title("Listening Time Trends by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Total Listening Time (minutes)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot 11: Scatter Plot of Listening Time
plt.figure(figsize=(12, 6))
plt.scatter(spotify_day_time.index, spotify_day_time["dailyListeningTime"], alpha=0.7, color='blue')
plt.title("Scatter Plot of Daily Listening Time")
plt.xlabel("Date")
plt.ylabel("Daily Listening Time (minutes)")
plt.grid()
plt.show()

# Plot 12: Monthly Variability in Listening Time
monthly_variability = spotify_day_time.groupby(spotify_day_time.index.month)["dailyListeningTime"].std()
plt.figure(figsize=(12, 6))
plt.plot(monthly_variability.index, monthly_variability, marker='o', color='red')
plt.title("Monthly Variability in Listening Time")
plt.xlabel("Month")
plt.ylabel("Standard Deviation of Listening Time")
plt.grid()
plt.show()

```
